```python
movie_ratings = ratings.groupby('movieId')['rating'].mean()

movie_ratings = movie_ratings.reset_index()

movies_df = movies.merge(movie_ratings, on='movieId')

movies_df.head()
```

```python
def recommend_by_genre(genre):

    results = movies_df[
        movies_df['genres'].str.contains(
            genre,
            case=False,
            na=False
        )
    ]

    results = results.sort_values(
        by='rating',
        ascending=False
    )

    return results[
        ['title', 'genres', 'rating']
    ].head(10)
```

```python
recommend_by_genre("Comedy")
```

```python
recommend_by_genre("Action")
```

```python
recommend_by_genre("Drama")
```

```python
def recommend_movies(genre, min_rating=4.0):

    results = movies_df[
        (movies_df['genres'].str.contains(
            genre,
            case=False,
            na=False
        ))
        &
        (movies_df['rating'] >= min_rating)
    ]

    results = results.sort_values(
        by='rating',
        ascending=False
    )

    return results[
        ['title','genres','rating']
    ].head(10)
```

```python
recommend_movies("Comedy", 4.5)
```

```python
recommend_movies("Action", 4.0)
```
